```python
import matplotlib.pyplot as plt
import numpy as np
import diagnostics
```

```python
exp = diagnostics.plume('../../data/Nyles/plume_ref4/', 'plume_ref4')
```

```python
exp.params
```

```python
variables = exp.read_vars(['z', 't'], '../../data/Nyles/plume_ref4/plume_ref4_00_hist.nc')
```

```python
z = variables['z']
t = variables['t']
```

### Mean horizontal buoyancy

```python
b_mean = exp.disk_average('b') 
```

```python
for k in range(0, b_mean.shape[0], 30):
    plt.plot(b_mean[k,:], z, label= f'{t[k]:.0f} s')
    
plt.legend()
plt.xlabel(r'Mean horizontal buoyancy ($m \ s^{-2}$)')
plt.ylabel('z (m)')
```

### Mean horizontal kinetic energy

```python
KE_mean = exp.disk_average('KE')
```

```python
for k in range(0, b_mean.shape[0], 30):
    plt.plot(KE_mean[k,:], z, label= f'{t[k]:.0f} s')
    
plt.legend()
plt.xlabel(r'Mean horizontal Kinetic Energy')
plt.ylabel('z (m)')
```

### $N^2$

```python
NN_mean = exp.disk_average('NN')

for k in range(0, b_mean.shape[0], 30):
    plt.plot(NN_mean[k,:], z[:-1], label= f'{t[k]:.0f} s')
    
plt.legend()
plt.xlabel(r'$N^2 $ ($s^{-2}$)')
plt.ylabel(r'z (m)')
```
